```python
# RNN is a type of deep learning neural network specially designed for sequential data or time series data where previous information is important .
# unlike traditional neural networks RNN has a memory mechanism . it remembers previous inputs and uses that information for predicting the next output
```

```python
# # working of RNN
# # in RNN the output from the previous step is paired again happily into the network along with the current input
# ex ==> "i love deep learning"

# # step 1 . --> read i
# step 2 .        remember i and read love
# step 3         remember i love and read deep
# step 4         continue similarity
# so previous context helps in understanding current data
```

```python
# architecture of RNN
# input ---> hidden state ----> output
# input ---> current state
# and the hidden state acts like memory
```

```python
# difference btw tradition neural network and rnn
```

```python
# feature               traditional neural network (ANN)                          RNN

# datatype              works on normal independent data                          works on sequential data

# memory                no memory                                                 has memory

# previous information  not used                                                  used

# input processing      all inputs independent                                    input processed sequentially

# best for              classification and regression                             nlp speech and sequence prediction

# connection type       forward only                                              feedback or recurrent connections

# context understanding| poor                                                      better

```

```python
# # important keypoint of rnn

# 1. rnn is good for sequence learning .
# 2. it uses hidden state for memory
# 3. same weights are reused at every time state
# 4. it can process variable length sequences
# 5. training is done using backpropagation for time(BPTT)
```

```python
# problem in basic RNN :::::

# 1. vanishing gradient problem ---> during backpropagation gradients become extremely small so the model cannot learn
# long term dependencies

# 2. exploading gradient problem ---> gradients become too large causing unstable training because of these problem
# RNN struggles with long sequences
```

```python
# disadvantage of rnn::::

# 1. slow training
# 2. vanishing gradient problem
# 3. difficult to train on long sequences
# 4. less Parallelization
```
