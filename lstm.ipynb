```python
# lstm(long short term memory )is a special type of rnn that is design to learn and remember information over long sequences
# this solves the vanishing gradient problem of traditional rnn by introducing a memory term and getting an architecture
```

```python
# # main introduction of rnn and
# an lstm cell consist of input xt and a previous hidden state ht-1 who contains information from the previous timesteps and acts as a
# short term memory and then previous cell state ct-1 it store long term information learn from previous state and it acts as a long
# term memory
```

```python

# # # lstm cell architecture
# # step 1 . forget gate ---> it decides what information should be remove from the cell state . if output is 0
# # which means forget completely and if output = 1 which means keep completely

# # 2. keep gate ---> it decide what u information should be stored if poutput is 0 means dont stor and if
# # output -=1 which means Store

# # 3. candidate memory ----> it create new candidate value that may be added to the memory and it generate new information
# # from current input

# # 4. update cell state ----> it update long term memory , it forget unncessary information , it add inportant new
# # information and it store result in new cell state

# # 5. output gate ----> it decide what information should be sent as output

# # 6. hidden state preparation ----> it produce the final output of current lstm cell
# cell state ---> it acts like a highway carrying inportant information through out  the sequence

# hidden state ----> it store the information needed immediately
```
