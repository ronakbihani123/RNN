```python
# Q1. what is difference between RNN and LSTM ?
```

```python
# RNN (Recurrent Neural Network) and LSTM (Long Short-Term Memory) are both neural networks designed to work with sequential data
# such as text, speech, and time-series data.

# Feature	                                            RNN	                                           LSTM
# Memory	                                Stores short-term information	                             Stores both short-term and long-term information
# Vanishing Gradient Problem	            Suffers from it	                                           Designed to overcome it
# Architecture	                          Simple recurrent structure	                               Contains memory cell and gates
# Training Speed	                        Faster	                                                   Slower due to complex structure
# Performance on Long Sequences         	Poor	                                                     Better
# Use Cases	                              Small sequences, simple tasks	                             NLP, translation, speech recognition, chatbots
```

```python
# Q2. what are the components of an LSTM cell ?
```

```python
# An LSTM consists of four key components: the Forget gate , Input gate , Cell gate and Output gate .
# the Forget gate decides what information to discard ,the input gate determines what new information to store,
# the cell state act as a long term memory and the output gate controls what information is passed to the next time step.
# these components helps LSTMs retain important information over long sequences and overcome the vanishing gradient
# problem that affects tradintional RNNs
```

```python
# Q3. why does lstm have 3 gates
```

```python
# LSTM has three gates because it needs to control the flow of information through its memory cell. These gates help the model decide:

# What information to forget
# What new information to store
# What information to output

# Without these gates, the model would either keep too much irrelevant information or lose important information.
```

```python
# 4. why cant a feed forward network handle sequence data effectively
```

```python
# A Feed Forward Neural Network (FFNN) cannot handle sequence data effectively
# because it treats every input independently and has no memory of previous inputs.

# Sequence data such as text, speech, and time-series data depends on the order and context of elements, which FFNNs cannot capture.
```

```python
# 5. if ur sequence length becomes very large thousand plus time steps what challenges may arise even
# in LSTM?
```

```python
# Although LSTMs are designed to handle long-term dependencies, very long sequences such as 1000+ time steps can still create challenges.
# These include difficulty remembering information from the distant past,
# residual vanishing gradient issues, increased training time due to sequential processing, higher memory consumption, and information
# bottlenecks in the cell state. Because of these limitations, Transformer-based architectures with self-attention
# are often preferred for very long sequences.
```

```python
# 6.  what will happen if the forget gate always ouput 0 and if the forget gate always output 1?              (most important )
```

```python
# If the Forget Gate always outputs 0, the LSTM completely discards the previous cell state at every time step and loses all long-term memory.
# If the Forget Gate always outputs 1, the LSTM never removes old information, causing the memory to accumulate irrelevant data over time.
# Therefore, the Forget Gate must learn values between 0 and 1 so that the network can selectively retain important information
# and forget unnecessary information.
```

```python
# 7. why is sigmoid used in LSTM gates ?
```

```python
# The sigmoid activation function is used in LSTM gates because it produces outputs between 0 and 1,
# which makes it ideal for controlling the flow of information.
```

```python
# 8. why is tanh used inside the lstm cell
```

```python
# Tanh (Hyperbolic Tangent) is used inside the LSTM cell because it produces values between -1 and +1,
# allowing the network to represent both positive and negative information while keeping values bounded.
```

```python
# 9. can lstm completely solve the vanishing gradient problem
```

```python
# LSTM reduces the vanishing gradient problem through its cell state and gates,
# but it cannot completely eliminate the problem, especially for extremely long sequences.
```

```python
# 10. why is word order important in nlp
```

```python
# Word order is important in NLP because changing the order of words can completely change the meaning of a sentence.

# Human language is sequential, and the meaning of a word often depends on the words that come before and after it.
```

```python
# 11.  how does lstm remember information from previous timesteps
```

```python
# An LSTM remembers information from previous time steps through its Cell State and Hidden State.

# The Cell State acts as a long-term memory that carries important information across the sequence,
# while the gates control what information is remembered, updated, or forgotten.
```

```python
# 12. what is bidirectional lstm? when would u use bi lstm ?
```

```python
# A Bidirectional LSTM is a type of LSTM that processes a sequence in both forward and backward directions. It consists of two LSTMs:
# one reads the sequence from left to right, while the other reads it from right to left. Their outputs are combined, allowing the model
# to use both past and future context. Bi-LSTMs are commonly used in tasks such as named entity recognition,
# sentiment analysis, POS tagging, and machine translation,
# where the entire sequence is available. However, they are not suitable for real-time prediction tasks because future inputs are not yet known.
```
